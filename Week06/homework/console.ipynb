What Python module **contains** the function to see if a file exists?
* The OS module contains a function that checks to see if a file exists. You can do `os.path.exists(file)`.

What is the function **within the module** that checks to see if the file exists?
* The function within the module is path.exists. 

What Python module **contains** the function to remove a file?
* The OS module also contains a function that can remove a file. 

What is the function **within the module** that removes the file?
* The function in the OS module to delete files is remove

What standard Python function is used to take **input** from a user prompt?
* The standard Python function that is used to take user input is the input function. 

```python
# This section will check to see if a file exists. If it does it will check with the user if they want to remove it or not.
from os import remove, path
# This is the file we are checking for.
file2check = "test.txt"
# This checks to see if the file exits
if path.exists(file2check):
    # The file does exist. This prompts the user if they want to delete it or not.
    decision = input("File found. Do you want to delete it? [y/n]: ")
    # Now lets check to see if the use wants to delete it.
    if decision.lower() == "y":
        # Delete the file
        print("You have decided to remove the file.")
        remove(file2check)
    elif decision.lower() == "n":
        # Keep the file
        print("You have decided to keep the file.")
    else:
        # No valid option was submitted.
        print("An invalid selection has been detected. Exiting...")
        exit
```

```python
from matplotlib.pyplot import hot
import paramiko
from getpass import getpass

# Create the prompt for the password
password = getpass(prompt="Please Enter your SSH password: ")

# Hist Information
host = "192.168.1.229"
port = 22
username = "ubuntu"


try: 
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    # Open the SSH session
    ssh.connect(host, port, username, password)
except paramiko.AuthenticationException:
    print("Authentication Failed.")
```

```python
# Start SFTP Session
from threading import local

# File to Upload
file = "fs.py"
# Start the SFTP Session
sftp = ssh.open_sftp()
# Copy the File
sftp.put(file, file)

# Command to run the file
command = "python3 fs.py -d /usr/bin > results.txt"
# Run the File 
ssh.exec_command(command)

# Remote File
remote_file = "results.txt"
# Lets get the results
sftp.get(remote_file, "results.txt")

# Close the SSH Session
ssh.close()
```
